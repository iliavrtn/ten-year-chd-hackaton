## Data Preparation 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score, confusion_matrix, accuracy_score, roc_curve
from framingham_score import *
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer

```

```python
def preprocess_data(df, handle_missings='median'):
    categorical_columns = ['is_smoking', "BPMeds", 'prevalentStroke', 'prevalentHyp', 'diabetes', 'sex', 'education']
    df.loc[df['is_smoking'] == "YES", 'is_smoking'] = 1
    df.loc[df['is_smoking'] == "NO", 'is_smoking'] = 0
    df.loc[df['sex'] == "M", 'sex'] = 1
    df.loc[df['sex'] == "F", 'sex'] = 0
    df["is_smoking"] = df["is_smoking"].astype(int)
    df["sex"] = df["sex"].astype(int)

    if handle_missings == 'median':
        for col in df.columns:
            if col not in categorical_columns:
                mean_col = df[col].median()
                df[col] = df[col].fillna(mean_col)
            else:
                mean_col = np.argmax(df[col])
                df[col] = df[col].fillna(mean_col)
    elif handle_missings == 'knn':
        for col in df.columns:
            if col not in categorical_columns:
                mean_col = df[col].median()
                df[col] = df[col].fillna(mean_col)

        imputator = KNNImputer(n_neighbors=5)
        df = pd.DataFrame(imputator.fit_transform(df), columns=df.columns)

    # Column to see if patient was in Hypertension during the sampling.
    df.insert(len(df.columns) - 1, 'isHyp', value=0)
    df.loc[(df['sysBP'] >= 130) | (df['diaBP'] >= 80), 'isHyp'] = 1

    # creating categorical column for packs of cigarettes.
    df.insert(len(df.columns) - 1, 'packsOfCigs', value=0)
    df.loc[df['cigsPerDay'] > 0, 'packsOfCigs'] = 1
    df.loc[df['cigsPerDay'] >= 10, 'packsOfCigs'] = 2
    df.loc[df['cigsPerDay'] >= 20, 'packsOfCigs'] = 3
    df.loc[df['cigsPerDay'] >= 30, 'packsOfCigs'] = 4

    # creating categorical column for glucose levels.
    df.insert(len(df.columns) - 1, 'glucose_level', value=0)
    df.loc[df['glucose'] > 2.6 * 18, 'glucose_level'] = 1
    df.loc[df['glucose'] > 4.7 * 18, 'glucose_level'] = 2
    df.loc[df['glucose'] > 6.3 * 18, 'glucose_level'] = 3
    df.loc[df['glucose'] > 8.5 * 18, 'glucose_level'] = 4

    # Score2 diabetes score (Framingham Risk Score)
    df.insert(len(df.columns) - 1, 'diabetes_score2', value=0)
    df['diabetes_score2'] = df.apply(lambda x: calculate_framingham_score(
        x['age'], x['sex'], x['is_smoking'], x['cigsPerDay'], x['BPMeds'], x['prevalentStroke'], x['prevalentHyp'],
        x['diabetes'],
        x['totChol'], x['sysBP'], x['diaBP'], x['BMI'], x['heartRate'], x['glucose'], x['education']), axis=1)
    return df.set_index('id')

def remove_outliers_iqr(df, columns, threshold=1.5):
    # Calculate the IQR for each column
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1

    # Define lower and upper bounds to identify outliers
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR

    # Filter rows without outliers
    mask = ~((df[columns] < lower_bound) | (df[columns] > upper_bound)).any(axis=1)
    return df[mask]


# Load Data
data_dir = r"C:\Users\soldier109\PycharmProjects\Moshal Medicine Hackaton 2023\Datasets\Cardiovascular Study Dataset"
trainset = preprocess_data(pd.read_csv(data_dir+"\\train.csv"), handle_missings='median')
```

## Handeling Outliers, Missing Values and Feature Selection

```python

# Define the columns you want to check for outliers
columns_to_check = trainset.drop(columns=['TenYearCHD']).columns

# Specify the IQR threshold (adjust as needed)
iqr_threshold = 2

# Remove rows with outliers
trainset_no_outliers = remove_outliers_iqr(trainset, columns_to_check, threshold=iqr_threshold)

# Print the shape before and after removing outliers
print(f"Original trainset shape: {trainset.shape}")
print(f"Trainset shape after removing outliers: {trainset_no_outliers.shape}")
```

```python
plt.figure(figsize=(15, 8))
correlation = trainset_no_outliers.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
```

After correlation some of the features were removed because of small variance of those features.

```python
print(trainset_no_outliers['TenYearCHD'].value_counts())
print(trainset['TenYearCHD'].value_counts())
```

Because of small amount of rows that have 1 in the target column, we will not remove outliers. We will use all the data we have.

```python
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, StandardScaler

X = trainset.drop(columns=['TenYearCHD']).copy()
y = trainset['TenYearCHD'].copy()
seed = 0
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)

print(X_train.shape)
print(X_test.shape)
```

## Logistic Regression

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
clf = LogisticRegression(fit_intercept=True, max_iter=10000)
clf.fit(X_train, y_train)
train_class_preds = clf.predict(X_train)
test_class_preds = clf.predict(X_test)

# Get the accuracy scores
train_accuracy = accuracy_score(train_class_preds, y_train)
test_accuracy = accuracy_score(test_class_preds, y_test)

print("The accuracy on train data is ", train_accuracy)
print("The accuracy on test data is ", test_accuracy)
```

```python
print(clf.coef_, clf.intercept_)
```

```python
# Get the confusion matrix for both train and test
plt.figure(figsize=(5, 3))
labels = ['Negative', 'Positive']
cm = confusion_matrix(y_train, train_class_preds)
print(cm)

ax = plt.subplot()
sns.heatmap(cm, annot=True, ax=ax, fmt='d')  #annot=True to annotate cells

# labels, title and ticks
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(labels)
ax.yaxis.set_ticklabels(labels)
```

```python
y_lr_predict_pro = clf.predict_proba(X_test)[:, 1]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_lr_predict_pro)
plt.figure(figsize=(5, 5))
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='Logistic Regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC curve')
plt.show()
```

## Handeling Data Imbalance

We can see that the imbalance is affecting out results heavily. Since most of our data have 0 in their target variable, the False Positive rate is large, since the model usually chooses 0 as the predicted label.
We'll try to tackle this problem using the Synthetic Minority Over-sampling Technique (SMOTE) algorithm,
using the k-neighbors strategy.

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)
```

```python
y_sm = pd.DataFrame(y_sm)
y_sm.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=0)
```

## Logistic Regression after handling imbalance

```python
clf = LogisticRegression(fit_intercept=True, max_iter=10000)
clf.fit(X_train, y_train)
train_class_preds = clf.predict(X_train)
test_class_preds = clf.predict(X_test)

# Get the accuracy scores
train_accuracy = accuracy_score(train_class_preds, y_train)
test_accuracy = accuracy_score(test_class_preds, y_test)

print("The accuracy on train data is ", train_accuracy)
print("The accuracy on test data is ", test_accuracy)
```

```python
print(clf.coef_, clf.intercept_)
```

```python
# Get the confusion matrix for both train and test
plt.figure(figsize=(5, 3))
labels = ['Negative', 'Positive']
cm = confusion_matrix(y_train, train_class_preds)
print(cm)

ax = plt.subplot()
sns.heatmap(cm, annot=True, ax=ax, fmt='d')  #annot=True to annotate cells

# labels, title and ticks
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(labels)
ax.yaxis.set_ticklabels(labels)
```

```python
y_lr_predict_pro = clf.predict_proba(X_test)[:, 1]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_lr_predict_pro)
plt.figure(figsize=(5, 5))
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='Logistic Regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC curve')
plt.show()
```

We can see that the rate of False Positive was alot better after using the SMOTE algorithm.

## Random Forest / GradientBoosting

```python
import time

t0 = time.time()
classifier = RandomForestClassifier()
grid_values = {'n_estimators': [100,175, 200,250], 'max_depth': [12, 15, 20, 25, 30]}
RF_GS_classifier = GridSearchCV(classifier, param_grid=grid_values, scoring='roc_auc', cv=5)

# Fit the object to train dataset
RF_GS_classifier.fit(X_train, y_train)
print(f'completed in {round(time.time() - t0, 2)} seconds')
```

```python
import time

t0 = time.time()
classifier = GradientBoostingClassifier()  # For GBM, use GradientBoostingClassifier()
grid_values = {'n_estimators': [100, 125, 150, 200,250], 'max_depth': [5, 7, 9, 12, 15],
               'learning_rate': [0.1, 0.05, 0.01]}
GBM_GS_classifier = GridSearchCV(classifier, param_grid=grid_values, scoring='roc_auc', cv=5)

# Fit the object to train dataset
GBM_GS_classifier.fit(X_train, y_train)
print(f'completed in {round(time.time() - t0, 2)} seconds')
```

```python
RF_bestvalues = RF_GS_classifier.best_params_
RF_GS_classifier.best_params_
```

```python
GBM_bestvalues = GBM_GS_classifier.best_params_
GBM_GS_classifier.best_params_
```

```python
rf_classifier = RandomForestClassifier(max_depth=RF_bestvalues['max_depth'], n_estimators=RF_bestvalues['n_estimators'])
rf_classifier.fit(X_train, y_train)
y_train_preds_rf = rf_classifier.predict(X_train)
y_test_preds_rf = rf_classifier.predict(X_test)

gbm_classifier = GradientBoostingClassifier(max_depth=GBM_bestvalues['max_depth'],
                                            n_estimators=GBM_bestvalues['n_estimators'],
                                            learning_rate=GBM_bestvalues['learning_rate'])

gbm_classifier.fit(X_train, y_train)
y_train_preds_gbm = gbm_classifier.predict(X_train)
y_test_preds_gbm = gbm_classifier.predict(X_test)
```

```python
print("Random Forest:\t", accuracy_score(y_train, y_train_preds_rf), accuracy_score(y_test, y_test_preds_rf))
print("Gradient Boosting:\t", accuracy_score(y_train, y_train_preds_rf), accuracy_score(y_test, y_test_preds_gbm))
```

We can see there isn't much of a difference between the models. We suspect the RF model to be slightly better.

## Random Forest Results

```python
y_predict_pro = rf_classifier.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_predict_pro)
print(roc_auc_score(y_test, y_predict_pro))
plt.figure(figsize=(5, 5))
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='Random Forest')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Random Forest ROC curve')
plt.show()
```
